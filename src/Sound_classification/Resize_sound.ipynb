```python
import os
from moviepy.editor import AudioFileClip, concatenate_audioclips
```

```python
# Define the paths
input_folder = "Bird_Recordings"
output_folder = "Short_clip"
target_duration = 10.0  # Desired duration in seconds

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# List all subdirectories in the input folder (each subdirectory represents a bird species)
bird_species_folders = [os.path.join(input_folder, folder) for folder in os.listdir(input_folder) if os.path.isdir(os.path.join(input_folder, folder))]

for bird_species_folder in bird_species_folders:
    # Get a list of MP3 files in the current bird species folder
    mp3_files = [f for f in os.listdir(bird_species_folder) if f.endswith(".mp3")]

    if mp3_files:
        for mp3_file in mp3_files:
            #print(f"Processing {mp3_file} in {bird_species_folder}")

            # Define the output folder for the current bird species
            species_output_folder = os.path.join(output_folder, os.path.basename(bird_species_folder))

            # Ensure the output folder for the bird species exists
            os.makedirs(species_output_folder, exist_ok=True)

            # Define the output file path within the bird species folder
            output_path = os.path.join(species_output_folder, mp3_file)

            try:
                # Load the original audio clip
                audio_clip = AudioFileClip(os.path.join(bird_species_folder, mp3_file))

                # Calculate the duration of the audio clip
                clip_duration = audio_clip.duration

                if clip_duration < target_duration:
                    # If the original audio is shorter than the target duration, repeat it
                    repeated_audio = audio_clip
                    while repeated_audio.duration < target_duration:
                        repeated_audio = concatenate_audioclips([repeated_audio, audio_clip])

                    # Trim the repeated audio to the target duration
                    repeated_audio = repeated_audio.subclip(0, target_duration)
                    repeated_audio.write_audiofile(output_path, codec='mp3')
                else:
                    print("lONGER", clip_duration)
                    # If the original audio is longer than the target duration, trim it
                    trimmed_audio = audio_clip.subclip(0, target_duration)
                    trimmed_audio.write_audiofile(output_path, codec='mp3')

            except Exception as e:
                print(f"An error occurred while processing {mp3_file}: {e}")
    else:
        print(f"No MP3 files found in {bird_species_folder}")
```
